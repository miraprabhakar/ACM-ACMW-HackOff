```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
rating = df['RATING'].tolist()
genre = df['GENRE'].tolist()

#print(f"\ncol_one_list:\n{genre}\ntype:{type(genre)}")
#print(f"\ncol_one_list:\n{rating}\ntype:{type(rating)}")
```

```python
matrix = []
for i in range(len(genre)):
    matrix.append([genre[i], rating[i]])

print(len(matrix))

def get_col(arr, col):
    return map(lambda x : x[col], arr)

result = []
for x in matrix:
    if x not in result:
        result.append(x)

#columns = list(zip(result))

#col1 = get_col(result, 0)
#print(col1)

col1 = [val[1] for val in result]
col2 = [val[0] for val in result]
#print(col1)
#print(col2)
```

```python
import matplotlib.pyplot as plt
plt.plot(col1,col2, ".")

def histogram_plot(x,
                   flag_save=True,
                   xlabel='rating',
                   ylabel='y',
                   lcolor='blue',
                   fs=14,fname='histogram_plot.pdf'):
        f, ax = plt.subplots(1,1,figsize=(5,5))
        ax.hist(col1,facecolor=lcolor,bins=100,edgecolor='black', alpha=0.05)
        ax.set_xlabel(xlabel,fontsize=fs)
        ax.set_ylabel(ylabel,fontsize=fs)

        if(flag_save):
            plt.savefig(fname,bbox_inches='tight',dpi=400)
histogram_plot(col1)
```
